## **Polynomial Fitting**

A ideia seria aproximar uma curva de uma função objetivo com polinômios da forma: $f(x) = a_0 + a_1x + ... + a_{n-1}x^{n-1}$. Ao realizar essa técnica, devemos nos atentar com overfitting (a função f se adaptar ao ruído dos dados) e com a maldição da dimensinalidade (com muitas features, o modelo pode se tornar muito complexo e necessitar de um número muito grande de amostrar para realizar o treinamento).

```python
# Bibliotecas
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
# Dados com ruído gaussiano
np.random.seed(0)
x = np.linspace(0, 11, 50)
noise = np.random.normal(0, 1, 50)
y = 3 * x + 5 + noise

fig = plt.figure(figsize=(8, 8))
plt.scatter(x, y, label='data')
plt.show()
```
